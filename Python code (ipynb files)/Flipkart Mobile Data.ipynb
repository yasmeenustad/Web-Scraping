```python
import pandas as pd
import numpy as np
from selenium import webdriver
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup
import requests
```

```python
Links = []
Name = []
MRP = []
Price = []
Ratings = []
Numbe_of_Ratings = []
Reviews = []

for k in range(16):
    url = f'https://www.flipkart.com/search?q=mobiles&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page={k}'
    req = requests.get(url)
    soup = BeautifulSoup(req.content)
    products = soup.find_all("div",class_='_2kHMtA')
    
    for i in products:
        links = i.find("a",class_='_1fQZEK')
        a = links.get('href')
        b = 'https://www.flipkart.com/'+a
        Links.append(b)

        names = i.find("div",class_='_4rR01T')
        name = names.text if names else ""
        Name.append(name)

        mrp = i.find('div', class_='_3I9_wc _27UcVY')
        mrps = mrp.text if mrp else ""
        MRP.append(mrps)

        price = i.find('div', class_='_30jeq3 _1_WHN1')
        prices = price.text if price else ""
        Price.append(prices)

        ratings = i.find('div', class_='_3LWZlK')
        ratings1 = ratings.text if ratings else ""
        Ratings.append(ratings1)

        reviews = i.find('span', class_='_2_R_DZ')
        reviews1 = reviews.text if reviews else ""
        Reviews_Ratings_count.append(reviews1)

        for j in Reviews_Ratings_count:
            Numbe_of_Ratings.append(j.split("&")[0].split(' ')[0])
            Reviews.append(j.split("&")[1].split(' ')[0].split("\xa0")[1])
            break
```

```python
data = {'Name':Name, 'MRP':MRP, 'Price':Price, 'Ratings':Ratings, 'Numbe_of_Ratings':Numbe_of_Ratings, 'Reviews':Reviews,'Links':Links}
```

```python
df = pd.DataFrame(data)
df
```
