```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from bs4 import BeautifulSoup
import requests

import time
from time import sleep
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
```

```python
driver=webdriver.Chrome("chromedriver.exe")
```

```python
base_url="https://www.swiggy.com"
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/112.0.0.0 Safari/537.36'}
driver.get(base_url)
driver.implicitly_wait(10)
```

```python
location = driver.find_element(By.XPATH , '//*[@id="location"]')
```

```python
location.send_keys('Bangalore, Karnataka, India')
```

```python
content = driver.page_source
```

```python
soup = BeautifulSoup(content)
soup
```

```python
len(soup.find_all('a',class_='_1j_Yo'))
```

```python
products = soup.find_all('div',class_='_3XX_A')
products
```

```python
links = []
for i in products:
    link1 = i.find('a',class_='_1j_Yo')
    link2 = link1.get('href') if link1 else ''
    links.append('https://www.swiggy.com'+link2)
    
print(links)
```

```python
Rest_Name = []
Ratings = []
Price_for_two = []
Cusines = []
Location = []
Delivery_review_no = []
URL = []

for j in links:
    headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/112.0.0.0 Safari/537.36'}
    req1 = requests.get(j, headers=headers)
    soup1 = BeautifulSoup(req1.text)
    
    name = soup1.find('p', class_='RestaurantNameAddress_name__2IaTv').text.strip()
    Rest_Name.append(name)
    
    rating = soup1.find('span',class_='RestaurantRatings_avgRating__1TOWY')
    ratings = rating.text.strip() if rating else ''
    Ratings.append(ratings)
    
    price = soup1.find('ul',class_='RestaurantTimeCost_wrapper__3YXF9')
    price1 = price.text.strip() if price else ''
    Price_for_two.append(price1)
    
    cusine = soup1.find('p',class_='RestaurantNameAddress_cuisines__mBHr2')
    cusine1 = cusine.text.strip() if cusine else ''
    Cusines.append(cusine1)

    location = soup1.find('p',class_='RestaurantNameAddress_area__2P9ib')
    location1 = location.text.strip() if location else ''
    Location.append(location1)
    
    review = soup1.find('span',class_='RestaurantRatings_totalRatings__3d6Zc')
    reviews = review.text.strip() if review else ''
    Delivery_review_no.append(reviews)
    
    URL.append('https://www.swiggy.com'+link2)
    
    time.sleep(0.01)
    
data = {'Rest_Name':Rest_Name, 'Ratings':Ratings,'Price_for_two':Price_for_two, 'Cusines':Cusines, 'Location':Location,
       'Delivery_review_no':Delivery_review_no, 'URL':URL}

df = pd.DataFrame(data)
```
