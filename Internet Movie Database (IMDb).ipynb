```python
import pandas as pd
import numpy as np
from selenium import webdriver
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup
import requests
```

```python
url = "https://www.imdb.com/chart/top/"
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/112.0.0.0 Safari/537.36'}
req = requests.get(url, headers=headers)
soup = BeautifulSoup(req.content)
```

```python
products = soup.find_all('li',class_='ipc-metadata-list-summary-item sc-bca49391-0 eypSaE cli-parent')
```

```python
Movie_Name = []
Year_of_Release = []
Rating = []
Links = []
Time_and_IMDb_Rating = []

for product in products:
    name = product.find('h3',class_='ipc-title__text')
    name1 = name.text if name else ''
    Movie_Name.append((' ').join(name1.split(' ')[1:]))

    year = product.find('span',class_='sc-b85248f1-6 bnDqKN cli-title-metadata-item')
    year1 = year.text if year else ''
    Year_of_Release.append(year1)
 
    ratings = product.find('span',class_='ipc-rating-star ipc-rating-star--base ipc-rating-star--imdb ratingGroup--imdb-rating')
    ratings1 = ratings.text if ratings else ''
    Rating.append(ratings1)
    
    imdb_rating = product.find('div',class_='sc-b85248f1-5 kZGNjY cli-title-metadata')
    list_ = imdb_rating.text
    Time_and_IMDb_Rating.append(list_[4:])
    
    link = product.find('a',class_='ipc-title-link-wrapper')
    Links.append(url+link.get('href'))
    
    
data = {'Movie_Name':Movie_Name, 'Year_of_Release':Year_of_Release, 'Rating':Rating, 'Time_and_IMDb_Rating': Time_and_IMDb_Rating, 
        'Links': Links}

IMDb_data = pd.DataFrame(data)
```

```python
# Title: "The Shawshank Redemption" is the title of the movie. It's the name by which the movie is known.

# Year of Release: "1994" is the year in which the movie was released. This indicates when the movie first became available for 
# audiences to watch.

# Duration: "2h 22m" represents the duration of the movie. In this case, it means the movie's runtime is 2 hours and 22 minutes.

# Rating: "R" is a movie rating. Movie ratings are assigned by film classification boards to indicate the appropriate audience 
# for a movie based on its content. "R" typically stands for "Restricted," which means the movie may contain content that is not 
# suitable for children under 17 without parental guidance.

# IMDb Rating: "9.3" is the IMDb (Internet Movie Database) rating of the movie. IMDb is a popular online database for movies, 
# TV shows, and other entertainment content. The IMDb rating represents the average user rating of the movie on a scale from 1 to
# 10.
```
